```python
from multiprocessing import Process
import numpy as np
```

```python
"""
Enter dataset size, number of CPU cores (threads), and interval to write to disk.  On larger data sizes,
this interval should be smaller ( > .001), as Python will urn out of memory otherwise.
"""

dataset = 100000000  # how many unique ids / rows for dataset
threads = 8  # how many threads to run, should equal cpu cores
flush_interval = int(dataset * .001)
interval = dataset / threads
start = 0
stop = interval
```

```python
def write_file(start, stop, i):
    
        my_line = ""
        print("Generating Data in memory for part {0}".format(i))
        count = stop - start

        # create income array
        income_avg, income_range  = 100000,20000
        income = np.random.normal(income_avg,income_range,count)

        # create age array
        age_avg, age_range = 40,20
        age = np.random.normal(age_avg,age_range,count)

        # create credit_score array
        credit_avg, credit_range = 740, 40
        credit = np.random.normal(credit_avg, credit_range, count)
        
        for y in range(count):
            my_line += "{0},{1},{2},{3}\n".format(str(start + y + 1), str(format(income[y], '.2f')), str(int(age[y])), str(int(credit[y])))
            if y % flush_interval == 0:
                print("At line {0} flushing to disk...".format(start + y + 1))
                with open("/Users/nate/Desktop/data/loan_applications_{0}.csv".format(i), 'a') as f:
                    f.write(my_line)
                my_line = ""

        print("Writing from {0} to {1} in loan_applications_{2}.csv\n".format(start, stop, i))
        with open("/Users/nate/Desktop/data/loan_applications_{0}.csv".format(i), 'a') as f:
            if my_line != "":
                f.write(my_line)
        print("Finished writing to loan_applications_{0}.csv\n".format(i))
```

```python
for i in range(threads):
    p = Process(target=write_file, args=(start, stop, i))
    p.start()
    start += interval
    stop += interval

p.join()
```
